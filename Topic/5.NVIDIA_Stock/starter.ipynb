## Data Loading

Load the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
df = pd.read_csv("data/NVDA_HistoricalData.csv")  # adjust filename if neededdf["Date"] = pd.to_datetime(df["Date"])df = df.set_index("Date").sort_index()
```

```python
# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Basic statistics
print(f"\nBasic statistics:")
df.describe()
```

```python
# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(df.index, df.iloc[:, 0], linewidth=1.5)
plt.title("Time Series Plot", fontsize=14, fontweight='bold')
plt.xlabel("Date", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
